```python
import numpy as np
import pandas as pd
```

```python
columns = [
    'Subject',
    'Measuring station',
    'repetition no.',
    'Neck girth',
    'Back neck point to waist',
    'Upper arm girth R',
    'Back neck point to wrist R',
    'Across back shoulder width',
    'Bust girth',
    'Waist girth',
    'Hip girth',
    'Thigh girth R',
    'Total crotch length',
    'Inside leg height',
]
```

```python
measures_data = pd.read_csv("data/MOVE4D/body_measurements.csv", sep=';')
measures_data = measures_data[columns]
station_list=np.append(measures_data["Measuring station"].unique(), "Meu")
```

```python
for column in columns:
    if column != 'Measuring station' and column != 'Subject':
        filtered = measures_data[column] > 2000
        measures_data[column].loc[filtered] /= 1000
```

```python
extra_data = pd.read_csv('data/MOVE4D/basic_informations.csv', sep=';')
```

```python
treated_data = pd.merge(measures_data, extra_data, on='Subject')
treated_data = treated_data.drop(['Pantone', 'BMI (kg/m^2)'], axis=1)
treated_data = treated_data[treated_data['Phase'] == 2]
```

```python
treated_data.to_pickle("data/move4d_cleaned_measures.zip")
```
