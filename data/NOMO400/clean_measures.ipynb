```python
import pandas as pd
import os
```

```python
df = pd.DataFrame()

for gender in ["male", "female"]:

    for file in os.listdir(f"{gender}m"):
        filetext = open(f"{gender}m/{file}").read()
        file_dataset = filetext.replace("MEASURE ", "").split("\n")[1:]
        file_dataset = list(map( lambda x: x.split("="), file_dataset))
        file_dataset.insert(0, ['gender', gender])
        file_dataset.insert(0, ['identifier', file[:-4]])
        file_datadict = dict(file_dataset[:-1])
        if gender == "female":
            file_datadict['CHEST_Circ'] = file_datadict["BUST_Circ"]
            del file_datadict['BUST_Circ']
        if gender == "male":
            file_datadict['NaturalWAIST_Circ'] = file_datadict["MaxWAIST_Circ"]
            del file_datadict['MaxWAIST_Circ']

        file_dataserie = pd.Series(file_datadict)
        df = pd.concat([df, file_dataserie], axis=1)
        
df = df.T
df.sort_values(by="identifier")
```

```python
df = df[['identifier', 'gender', 'CHEST_Circ', 'NaturalWAIST_Circ', 'HIP_Circ', 'Thigh_Circ', 'Bicep_Circ', 'Neck_Circ', 'Head_Top_Height']]
df.sort_values(by="identifier").reset_index(drop=True).to_pickle("measures.zip")
```
